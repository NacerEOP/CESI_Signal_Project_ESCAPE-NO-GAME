# Rappel du contexte
L'agence AIL3C, spécialisée dans la lutte contre la cybercriminalité, développe des technologies pour aider ses agents. Actuellement, nous travaillons sur un projet suite à la perte de l'agent K57, qui a été piégé dans une salle sans réseau pendant une mission. Cette situation met en évidence le besoin d'une solution pour permettre la communication dans de telles conditions.

# Traitement de la problématique
L'incident de l'agent K57 met en évidence une problématique critique : **l'absence de communication dans un environnement sans réseau**. L'agent, piégé dans une salle sans accès à un réseau sans fil, n'a pas pu envoyer de signal de détresse. Cette situation souligne **le besoin d'une solution permettant de maintenir une communication même dans des lieux isolés où les technologies modernes (Wi-Fi, Bluetooth, etc.) ne fonctionnent pas.**

La problématique est donc **de trouver une méthode fiable pour permettre la communication dans des environnements où les réseaux sans fil sont inaccessibles, en utilisant des équipements disponibles sur place ou des technologies alternatives.**

# Presentation et explication de la proposition

## Notre proposition:
- **Utiliser l'équipement qui était à disposition dans la salle (microphone d'audioconférence et pad miniature spécial espion de l'agent) pour transmettre un message de détresse via le câblage du bâtiment qui est mis sous écoute par l'agence.**

Le microphone est un appareil conçu pour convertir les ondes sonores en signaux électriques. Ce processus se déroule en plusieurs étapes distinctes :

- **Réception des ondes sonores**:
Les ondes sonores, qui sont des variations de pression dans l’air, entrent dans le microphone et frappent un diaphragme, une membrane très fine et souple. Ce diaphragme vibre en réponse aux fluctuations de pression des ondes sonores.

- **Conversion des vibrations en signal électrique**:
Selon le type de microphone, la conversion des vibrations en signal électrique se fait de manière différente :

-Microphone dynamique : Le diaphragme est attaché à une bobine placée dans un champ magnétique. Lorsque le diaphragme vibre, la bobine se déplace dans le champ magnétique, générant un courant électrique.

-Microphone à condensateur : Le diaphragme fait partie d’un condensateur, dont l’écart entre les plaques change avec les vibrations. Ce changement modifie la capacité du condensateur et produit une variation de tension.

-Microphone piézoélectrique : Un cristal piézoélectrique se déforme sous l’effet des vibrations, générant ainsi un signal électrique proportionnel à ces déformations.
- **Amplification et transmission du signal**:
Le signal électrique produit est généralement faible et nécessite d’être amplifié avant d’être envoyé à un appareil d’enregistrement ou de diffusion, comme une table de mixage, un ordinateur ou un amplificateur.

- **Traitement du signal**:
Une fois amplifié, le signal peut être traité pour être enregistré, modifié ou diffusé, en fonction de l’application souhaitée.
### Caractéristiques d'une onde:

    <img src="Onde.jpg" style="width:60%;height:auto;">
</div>
Une onde est une perturbation qui se propage dans un milieu, transmettant de l'énergie sans transporter de matière. Elle possède plusieurs caractéristiques fondamentales qui définissent son comportement et son interaction avec l'environnement. Ces caractéristiques incluent :

- **Amplitude**:
L'amplitude d'une onde représente l'intensité de la perturbation, c'est-à-dire l'écart maximal entre la position d'équilibre et la crête (ou le creux) de l'onde. Une grande amplitude correspond à une onde plus énergique, souvent perçue comme plus forte ou plus bruyante dans le cas des ondes sonores.

- **Fréquence**:
La fréquence désigne le nombre de cycles complets (oscillations) qu'une onde effectue en une seconde, mesurée en Hertz (Hz). Une fréquence élevée signifie que l'onde oscille rapidement, ce qui peut être associé à des sons aigus ou des hautes fréquences dans le cas des ondes sonores.

- **Longueur d'onde**:
La longueur d'onde est la distance entre deux points successifs de même phase sur l'onde, par exemple, entre deux crêtes ou deux creux. Elle est inversement proportionnelle à la fréquence : plus la fréquence est élevée, plus la longueur d'onde est courte.

- **Vitesse de propagation**:
La vitesse de propagation d'une onde est la vitesse à laquelle l'onde se déplace à travers le milieu. Elle dépend des propriétés du milieu traversé, comme sa densité ou sa rigidité. Par exemple, les ondes sonores se propagent plus vite dans l'eau que dans l'air.

- **Période**:
La période est le temps nécessaire pour qu'un cycle complet de l'onde se produise. Elle est l'inverse de la fréquence : une fréquence élevée correspond à une période courte et vice versa.

- **Phase**:
La phase décrit la position d'un point donné sur l'onde à un instant précis. Elle permet de décrire l'état de l'onde dans son cycle, ce qui est particulièrement important pour analyser les interférences entre plusieurs ondes.

- **Propagation**:
Les ondes peuvent se propager de différentes manières, selon leur nature. Les ondes mécaniques, comme les ondes sonores, nécessitent un milieu matériel pour se propager, tandis que les ondes électromagnétiques, comme la lumière, peuvent se propager dans le vide.

- **Polarisation-**:
Cette caractéristique s'applique aux ondes transversales, comme les ondes lumineuses. La polarisation fait référence à l'orientation de l'oscillation de l'onde par rapport à sa direction de propagation.

## Pourquoi un signal haute fréquence ?
- Pour éviter que le signal de détresse ne soit entendu par l'ennemi à travers le système d'audioconférence du bâtiment, le signal doit être intercepté uniquement par l'équipement de l'agence. Les ondes haute fréquence sont inaudibles pour l'oreille humaine car elles dépassent la plage de fréquences que notre système auditif peut détecter. Cela est dû à la limitation des récepteurs dans l'oreille humaine, qui ne peuvent capter que des vibrations dans la plage de 20 Hz à 20 kHz.

- De plus, une onde haute fréquence possède davantage d'énergie, ce qui lui permet de traverser des distances plus longues avec moins de perte de puissance. Cela la rend idéale pour des communications discrètes et efficaces, même sur de grandes distances. Les ondes au-delà de cette plage, comme les ultrasons, sont invisibles et inaudibles pour l'humain, mais peuvent être utilisées pour des applications spécifiques, telles que la communication discrète.
