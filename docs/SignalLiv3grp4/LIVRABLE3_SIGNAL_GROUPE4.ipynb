## <span style="color: #FF6347;">1. Introduction</span>

L'objectif de ce livrable est de décrire en détail le fonctionnement de la chaîne de transmission permettant la communication d'un agent infiltré avec l'extérieur à travers un signal ultrasonique. Nous allons modéliser chaque étape de transmission, justifier nos choix techniques et quantifier les paramètres clés.


## <span style="color: #FF6347;">2. Schéma de la Chaîne de Transmission</span>

Voici la structure détaillée de la transmission du signal depuis la génération jusqu’à la réception et son exploitation :

1. **Génération du Message (Pad de l'Agent)**
   - Le message texte (par exemple, un caractère ASCII comme "A") est converti en binaire.
   - Chaque caractère est traduit en une séquence binaire de 8 bits.
   - Cette séquence binaire est ensuite utilisée pour générer un signal modulaire.

2. **Modulation Ultrasonique (Conversion en Onde Sonore)**
   - Le signal binaire est modulé par une **modulation de fréquence (FSK)**.
     - **0** → Fréquence de **24.9 kHz** (exemple)
     - **1** → Fréquence de **25.1 kHz** (exemple)
   - Cela donne une onde ultrasonore qui peut être émise par le pad de l'agent.

3. **Transmission du Signal via Système Audioconférence**
   - Le signal modifié est transmis à travers le système d'audioconférence du bâtiment. Ce signal est une onde sonore dans la plage des ultrasons (25 kHz).
   - Le support de transmission utilisé ici est l’air, avec la possibilité de rencontrer des interférences dues aux structures physiques du bâtiment.

4. **Réception du Signal (Microphone du Système d'Écoute)**
   - Le signal ultrasonore est capté par un microphone relié au système de surveillance.
   - Le microphone reçoit le signal en analogique, qui est ensuite converti en un signal numérique par un convertisseur analogique-numérique (ADC).

5. **Filtrage Numérique (Isolation des Fréquences Utiles)**
   - Un **filtre passe-bande** est appliqué pour isoler les fréquences de 24.9 kHz à 25.1 kHz.
   - Cela permet de supprimer les autres fréquences parasites et de ne conserver que le signal utile.

6. **Démodulation (Extraction du Message Binaire)**
   - Le signal filtré est analysé pour identifier les fréquences portantes (24.9 kHz pour le bit 0 et 25.1 kHz pour le bit 1).
   - À partir de ces fréquences, on reconstruit la séquence binaire initiale.

7. **Reconstruction du Message (Conversion du Binaire en Texte)**
   - La séquence binaire obtenue est ensuite convertie en texte ASCII pour obtenir le message initial envoyé par l’agent.

## <span style="color: #FF6347;">3. Description des Étapes</span>

### **3.1. Génération du Signal**

Le processus commence par la génération du message. Ce message est d'abord converti en une séquence binaire en utilisant le standard ASCII. Chaque caractère (par exemple, le caractère "A") est converti en une séquence de bits.

Une fois que nous avons la séquence binaire, elle est ensuite utilisée pour moduler un signal porteur. Pour la modulation, nous utilisons la méthode de **modulation par déplacement de fréquence (FSK)**, qui est idéale pour transmettre des données numériques par onde sonore.

Dans cette modulation, la fréquence du signal porteuse varie en fonction du bit à transmettre. Par exemple :
- **0** est représenté par une fréquence de **24.9 kHz**.
- **1** est représenté par une fréquence de **25.1 kHz**.

L’objectif ici est de garantir que le signal émis reste inaudible à l'oreille humaine tout en étant suffisamment détectable par le système de surveillance.



### **3.2. Transmission du Signal**

Une fois que le signal est modifié selon la modulation FSK, il est transmis via le **système d’audioconférence** du bâtiment. Ce système transporte l’onde sonore à travers l’air, en utilisant l’acoustique pour véhiculer l’information.

Le signal ultrasonore peut rencontrer des obstacles physiques qui atténuent sa propagation. Cependant, l'objectif est que, même avec une certaine perte d’intensité, le signal reste détectable par le microphone du système d’écoute.

Le choix d’utiliser un signal ultrasonore est basé sur sa capacité à traverser des structures physiques sans causer de bruit détectable pour un observateur extérieur. Cette transmission utilise l'air comme support, mais des interférences possibles peuvent affecter la qualité du signal.


### **3.3. Réception et Filtrage Numérique**

Le signal est capté par un microphone placé à proximité, qui capte les ondes ultrasonores. Ces ondes sonores sont ensuite converties en signal numérique par un **convertisseur analogique-numérique (ADC)**.

Une fois le signal converti en numérique, il est nécessaire de le filtrer afin d’isoler la fréquence d’intérêt et d'éliminer les autres fréquences parasites. Pour ce faire, un **filtre passe-bande** est appliqué.

Le **filtre passe-bande numérique** est conçu pour ne laisser passer que la plage de fréquences correspondant à celle utilisée pour la transmission (autour de 25 kHz). Il permet ainsi de séparer le signal utile des bruits et autres interférences. Ce processus est crucial pour s'assurer que seuls les bits du message encodé soient extraits et non des fréquences indésirables.


### **3.4. Démodulation et Reconstruction du Message**

Une fois que le signal a été filtré, il est nécessaire de le démouler pour récupérer les bits d'origine. Cette étape consiste à analyser les variations de fréquence du signal pour déterminer si chaque bit correspond à une fréquence de 24.9 kHz (pour un 0) ou de 25.1 kHz (pour un 1).

Le signal est ensuite converti en une nouvelle séquence binaire, qui est finalement transformée en texte. Ce texte représente le message initialement envoyé par l'agent.


## <span style="color: #FF6347;">4. Quantification des Paramètres et Calculs</span>

### 4.1. Calcul du Temps de Transmission

Le temps de transmission pour un caractère ASCII (8 bits) est donné par :

$$
T = \frac{8}{bode}
$$

avec \( bode \) le débit binaire ou la vitesse de transmission (en bits par seconde, bps).

Si la modulation utilise un débit de 300 bps et que chaque bit est envoyé sur un cycle complet :

$$
bode = 300 \text{ bits/s}
$$

Alors, le temps pour transmettre un caractère est :

$$
T = \frac{8}{300} = 0.0266 \text{ s} = 26.66 \text{ ms}
$$

Pour une phrase de 10 caractères, le temps total sera :

$$
T_{total} = 10 \times 26.66 = 266.6 \text{ ms}
$$

3. Transmission par audio.
